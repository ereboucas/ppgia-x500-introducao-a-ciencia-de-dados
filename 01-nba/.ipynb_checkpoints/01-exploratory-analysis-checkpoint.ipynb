# Split htwt into Height and Weigth converted to International System of units

```python
import numpy as np

df[['height_ft', 'weight_lbs']] = df['htwt'].str.split(',', expand=True)

def convert_height_to_centimeters(height):
    if pd.isna(height) or height.strip() == '':
        return np.nan
    feet, inches = height.split("'")
    inches = inches.strip('"').strip()
    total_inches = int(feet.strip()) * 12 + int(inches)
    meters = total_inches * 2.54
    return round(meters, 0)

def convert_weight_to_kg(weight):
    if pd.isna(weight) or weight.strip() == '':
        return np.nan
    pounds = int(weight.strip('lbs').strip())
    kg = pounds * 0.453592
    return round(kg, 0)

df['height_cm'] = df['height_ft'].apply(convert_height_to_centimeters)
df['weight_kg'] = df['weight_lbs'].apply(convert_weight_to_kg)

df.groupby('name').head(1)[df['htwt'].notna()][['name', 'height_cm', 'weight_kg']]
```
