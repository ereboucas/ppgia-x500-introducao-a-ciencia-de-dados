# Housing

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('housing.csv')
```

# Data cleaning

```python
df = df.dropna(subset=['total_bedrooms'])
```

```python
df['ocean_proximity_numeric'] = df['ocean_proximity'].map({'OCEAN': 0, 'INLAND': 1, 'ISLAND': 2, 'NEAR BAY': 3, 'NEAR OCEAN': 4})
df['color'] = df['ocean_proximity'].map({'OCEAN': 'blue', 'INLAND': 'brown', 'ISLAND': 'cyan', 'NEAR BAY': 'red', 'NEAR OCEAN': 'green'})
```

```python
df.plot(kind='scatter', x='latitude', y='longitude', c='color', s=1, alpha=0.2)
```
